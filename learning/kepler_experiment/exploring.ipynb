```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

## Misc. First Plots

```python
raw_df = pd.read_csv('https://raw.githubusercontent.com/HarritonResearchLab/sunnyhills/main/learning/kepler_experiment/eu_exoplanet_catalog.csv')
```

```python
plt.scatter(raw_df['orbital_period'], raw_df['semi_major_axis'])
#plt.xlim(0, 8)
#plt.ylim(0,300)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Orbital Period')
plt.ylabel('Orbital Distance')

plt.show()
```

```python
# Create now a sky map of the results

# Set a dark background (the night sky is ... dark)
plt.style.use('ggplot')

# Create a figure and then apply the aitoff projection
plt.figure(figsize=(12, 8))
plt.subplot(projection="aitoff")


# Replace the standard x ticks (longitude) with the ecliptic coordinates
plt.xticks(ticks=np.radians([-150, -120, -90, -60, -30, 0, \
                             30, 60, 90, 120, 150]),
           labels=['150', '120', '90', '60', '30', '0', \
                   '330', '300', '270', '240', '210'])

# Set the axes labels
plt.xlabel('Eclip. long. in deg')
plt.ylabel('Eclip. lat. in deg')

# Create a legend and grid
plt.legend()
plt.grid(True)

plt.show()
```
